```python
# **Q-11** Total Charges should be numeric amount. Let's convert it to numerical data type and check whether it has any null values or not.
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.isnull().sum()
```

```python
# **Q-12** As we can see there are 11 missing values in TotalCharges column. Let's check these records
missing_total_charges=df[df['TotalCharges'].isnull()]
missing_total_charges
```

```python
# **Q-13** Convert the target variable 'Churn' in a binary numeric variable i.e. Yes=1 ; No = 0
df['Churn']=df['Churn'].map({'yes':1, 'No':0})
df['Churn']
```

```python
# **Q-14** Build a corelation of all predictors with 'Churn' and plot a bar chart (Use dataframe.corr() function)
# # Convert 'TotalCharges' to numeric, handling errors
# df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# # If 'Churn' is not numeric, convert it (e.g., if it's 'Yes'/'No')
# if df['Churn'].dtype == object:
#     df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
numerical_features = df.select_dtypes(include=np.number).columns
correlation = df[numerical_features].corr()['Churn'].drop('Churn')
plt.figure(figsize=(10, 6))
correlation.plot(kind='bar')
plt.title('Correlation of Predictors with Churn')
plt.xlabel('Predictors')
plt.ylabel('Correlation')
plt.xticks(rotation=45, ha='right')
plt.show()
print('correlation')
```

```python
# **Q-15** Plot a pie chart showing the distribution of the Payment Methods

df['PaymentMethod'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Payment Methods')
plt.ylabel('')
plt.show()
```

```python
# **Q-16** Plot a line chart showing the distribution of tenure
tenure_counts=df['tenure'].value_counts().sort_index()
plt.figure(figsize=(10,6),)
plt.plot(tenure_counts.index, tenure_counts.values)
plt.title('Distribution of Tenure')
plt.xlabel('tenure')
plt.ylabel('count')
plt.grid(True)
plt.show()
```
